```python
import matplotlib.pyplot as plt
import numpy as np
y = np.random.randint(5,30,10)
print(y)
plt.pie(y)
plt.show()
```

```python
#Histogram
y = np.random.randint(5,30,100)
plt.hist(y)
plt.show()
```

```python
y = np.random.randint(5,30,10)
x = np.arange(0,10)
plt.bar(x,y)
plt.show()
plt.barh(x,y)
plt.show()
```

```python
y = np.random.randint(5,30,500)
x = np.arange(0,500)
plt.scatter(x,y)
plt.show()
plt.hexbin(x,y)
plt.show()
```

```python
x = np.arange(0,20)
y = np.random.randint(5,30,20)
plt.title("Class Performance")
plt.xlabel("Class Tests")
plt.ylabel("Marks Obtained")
plt.show()
#plt.plot(x,y,'g', marker = 'p', ms = 15, mfc = 'r')
#plt.show()
plt.plot(x,y,':d', marker = 'd', ms = 10, mfc = "r")
plt.show()
```
